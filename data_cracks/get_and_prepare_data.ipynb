# 2.) Train Test Split

```python
import os
import glob
import random

random.seed(42) # Fix random seed

TESTSPLIT = 0.20 # We keep 20% of the data for testing

# Create training and test directory
for f in ["./train", "./test"]:
    if os.path.exists(f):
        os.system(f"rm {f}")
    os.system(f"mkdir -p {f}/Positive")
    os.system(f"mkdir -p {f}/Negative")
```

```python
# Check if folder exists
assert os.path.exists("./Negative"), "Folder does not exist! Please check if you followed all steps from part 1.)"
assert os.path.exists("./Positive"), "Folder does not exist! Please check if you followed all steps from part 1.)"
```

```python
# Perform test/train split
filenames = glob.glob('*/*.jpg')
filenames = sorted(filenames, key=lambda k: random.random()) # Shuffle data

for file in filenames:
    
    classpath, name = os.path.split(file)
    
    folder = "./train" if random.random() > TESTSPLIT else "./test"
        
    fileDest = os.path.join(folder, classpath+"/", name)
    
    os.system(f"cp {file} {fileDest} ")

    print(f"Processing {file}")
```
